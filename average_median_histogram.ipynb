```python
import pandas as pd

# read data
data = pd.read_csv("data_clean.csv")
```

```python
from statistics import mean

def compute_freq_per_n(lst):
    amount_freq = {}
    for item in lst:
        if item in amount_freq:
            amount_freq[item] += 1
        else:
            amount_freq[item] = 1
            
    amount = [x for x in amount_freq]
    amount = sorted(amount)
    
    freq = []
    for n in amount:
        freq.append(amount_freq[n])
        
    return amount, freq

amount_killed, frequency_killed = compute_freq_per_n(data["n_killed"])
amount_injured, frequency_injured = compute_freq_per_n(data["n_injured"])
        
    
    
print(amount_killed)
print(frequency_killed)
print(amount_injured)
print(frequency_injured)

average_killed = mean(data["n_killed"])
average_injured = mean(data["n_injured"])
    

print("Total people killed:", sum(data["n_killed"]))
print("Average people killed:", average_killed)
print("Total people injured:", sum(data["n_injured"]))
print("Average people injured:", average_injured)

```

```python
from bokeh.io import show, output_notebook
from bokeh.plotting import figure
import math

output_notebook()

def make_histogram(x_axis,y_axis,log,h,w,tit):
    x_str = [str(x) for x in x_axis]
    if log == True:
        y_axis_used = [math.log10(x) for x in y_axis]
    else:
        y_axis_used = y_axis
        
    p = figure(x_range=x_str, plot_height=h, title=tit)
    p.vbar(x=x_str, top=y_axis_used, width=w)

    p.xgrid.grid_line_color = None
    p.y_range.start = 0

    show(p)
    
make_histogram(amount_killed,frequency_killed,True,500,0.2,"Murders")
make_histogram(amount_injured,frequency_injured,True,500,0.2,"Murders")
```
